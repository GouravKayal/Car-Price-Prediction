```python
import re
def parse_power_column(value):
    if not isinstance(value, str):
        return None, None

    try:
        # Extract power and RPM part
        power_match = re.search(r"([\d.]+)\s*(ps|bhp|hp|kW)", value, re.IGNORECASE)
        rpm_match = re.search(r"@.*?(\d+)(?:-(\d+))?rpm", value, re.IGNORECASE)

        if not power_match:
            return None, None

        power = float(power_match.group(1))
        unit = power_match.group(2).lower()

        # Convert to PS only if unit is bhp or hp
        if unit == "bhp":
            power *= 1.01387
        elif unit == "hp":
            power *= 1.014  # Approx conversion
        elif unit == "kW":
            power *= 1.0
        # Extract RPM
        if rpm_match:
            if rpm_match.group(2):  # If it's a range (e.g., 5000-6500rpm)
                rpm = int(rpm_match.group(2))
            else:
                rpm = int(rpm_match.group(1))
        else:
            rpm = None

        return round(power, 2), rpm

    except:
        return None, None
```

```python
df[['Max_Power_PS', 'POWER_RPM']] = df['Power'].apply(lambda x: pd.Series(parse_power_column(x)))
```

```python

def parse_torque_column(value):
    if not isinstance(value, str):
        return None, None

    try:
        # Extract torque and RPM part
        torque_match = re.search(r"([\d.]+)\s*(nm)", value, re.IGNORECASE)
        rpm_match = re.search(r"@.*?(\d+)(?:-(\d+))?rpm", value, re.IGNORECASE)

        if not torque_match:
            return None, None

        torque = float(torque_match.group(1))

        # Extract RPM
        if rpm_match:
            if rpm_match.group(2):  # If it's a range (e.g., 1500-2500rpm)
                rpm = int(rpm_match.group(2))
            else:
                rpm = int(rpm_match.group(1))
        else:
            rpm = None

        return round(torque, 2), rpm

    except:
        return None, None
df[['Max_Torque_Nm', 'TORQUE_RPM']] = df['Torque'].apply(lambda x: pd.Series(parse_torque_column(x)))
```

```python
#Using K-Nearest Neighbours Algorithm
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Initialize and train the K-Nearest Neighbors Regressor model
knn = KNeighborsRegressor(n_neighbors=4) # You can adjust the number of neighbors (k)
knn.fit(X_train, y_train)

# Make predictions on the scaled training and test sets
y_pred_train_knn = knn.predict(X_train)
y_pred_test_knn = knn.predict(X_test)

# Evaluate the model
mse_train_knn = mean_squared_error(y_train, y_pred_train_knn)
rmse_train_knn = np.sqrt(mse_train_knn)
r2_train_knn = r2_score(y_train, y_pred_train_knn)

mse_test_knn = mean_squared_error(y_test, y_pred_test_knn)
rmse_test_knn = np.sqrt(mse_test_knn)
r2_test_knn = r2_score(y_test, y_pred_test_knn)

print(f'K-Nearest Neighbors Regressor - Train Set RMSE: {rmse_train_knn}')
print(f'K-Nearest Neighbors Regressor - Train Set R2: {r2_train_knn}')
print(f'K-Nearest Neighbors Regressor - Test Set RMSE: {rmse_test_knn}')
print(f'K-Nearest Neighbors Regressor - Test Set R2: {r2_test_knn}')
```
